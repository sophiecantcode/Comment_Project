```python
# we made our own csv file reading from our list of names, final grade, project grades, and then, on a scale of 1-5, participation, collaboration,
# focus, enthusiasm, and love for the code
names = ['Stacy Johnson', 'William Little', 'Joanna Smith', 'Sofia Cinnamon', 'Elizabeth Finn']
final_grades = ['103', '90', '80', '75','53']
project_grades = ['98', '70', '95', '73', '45']
participation = ['4', '5', '3', '3', '2']
collaboration = ['4', '5', '4', '4', '1']
focus = ['5', '3', '4', '3', '1']
enthusiasm = ['5', '3', '2', '2', '1']
loveforthecode = ['5', '3', '2', '2', '1']
gender = ['g','b','g', 'g', 'nb']
import csv
with open('teachers.csv', 'w') as file:
    myfile = csv.writer(file)
    myfile.writerow(['Name','Final Grade', 'Project Grade', 'Participation', 'Collaboration', 'Focus', 'Enthusiasm', 'Love for the Code', 'Attendance'])
    for i in range(0,5):
        myfile.writerow([names[i], final_grades[i], project_grades[i], participation[i], collaboration[i], focus[i], enthusiasm[i], loveforthecode[i], gender[i]])

with open('teachers.csv') as file:
    for row in file:
        print(row)
```

```python
# we made functions for each of the components of the csv file above that added a few sentences to our overall string to describe the grade of the
# student for that particular skillset and any advice that we might give for the next semester
def final_grade_function(final_grade, string):
    '''adds a few sentences to the preexisting string about the student's final grade in the class and advice about the next semester'''
    if final_grade >= 93:
        string = string + f"    {name} did very well this semester — {gender[0]} ended with a final grade of {final_grade} which is an A."
    if 90 <= final_grade < 93:
        string = string + f"    {name} did well this semester — {gender[0]} ended with a final grade of {final_grade} which is an A-."
    if 80<= final_grade < 90:
        string = string + f"    {name} did pretty well this semester — {gender[0]} ended with a final grade of {final_grade} which is in the B range. I think that {name} should meet with me more often and ask more questions in class."
    if 70 <= final_grade <80:
        string = string + f"    {name} ended the semester with a final grade of {final_grade} which is in the C range — {gender[0]} should try to meet with me weekly and ask questions about the material."
    if final_grade < 70:
        string = string + f"    {name} ended the semester with a final grade of {final_grade} which is below the C range — {gender[0]} should try to meet with me weekly and ask questions about the material. We might need to reassess {name}'s place in the class."
    return string

def project_grade_function(project_grade, string):
    '''adds a few sentences to the preexisting string about the student's project grade in the class and advice about the next semester'''
    if project_grade >= 95:
        string = string + f" For this semester, {gender[0]} received an overall project grade of {project_grade}. This score is one of the best in the class and demonstrates {gender[1]} proficiency across a wide range of projects, including the Banking Project and the Teacher Commenting Project.\n\n"
    if 85 <= project_grade < 95:
        string = string + f" For this semester, {gender[0]} received an overall project grade of {project_grade}. This score demonstrates a consistently high level of work from {gender[1]}  with small areas in which some growth may benefit {name}'s coding.\n\n"
    if 75 <= project_grade < 85:
        string = string + f" For this semester, {gender[0]} received an overall project grade of {project_grade}. This score offers room for improvement across a few areas of {gender[2]} code and shows that {name} has a bit of difficulty when it comes to projects. Hopefully, through more frequent meetings with me in the second semester, this grade will go up!\n\n"
    if project_grade < 75:
        string = string + f" For this semester, {gender[0]} received an overall project grade of {project_grade}. I would strongly encourage {name} to ask more questions when unsure of how to approach {gender[1]} code. Furthermore, increased levels of effort both in class and outside of class would most likely result in a higher project grade next semester.\n\n"
    return string

def participation_function(participation, string):
    '''adds a few sentences to the preexisting string about the student's participation level in the class'''
    if participation == '5':
        string = string + f"    For {gender[2]}, participating in class is not an issue at all, in fact it is a strength! Whenever I look {gender[1]} way, {gender[1]} hand is up! I have never ever been  in class and seen {gender[0]} hand not waving about in the air. I love the participation!"
    if participation == '4':
        string = string + f"    For {gender[2]}, participating in class is not an issue! Without overpowering {gender[1]} peers, {name} offers {gender[1]} insights and contributes to class discussions in a meaningful and enlightening way."
    if participation == '3':
        string = string + f"    For {gender[2]}, participating in class rarely an issue! Sometime {gender[0]} seems a little bit shy or hesitant to offer up {gender[1]} ideas, but when {gender[0]} raises {gender[1]} hand, {name} always contributes something meaningful to class discussions."
    if participation == '2':
        string = string + f"    For {gender[2]}, participating in class is something to work on. In the upcoming semester, I would love for {name} to take more risks and raise {gender[1]} hand more with all those great thoughts!"
    if participation == '1':
        string = string + f"    It would be great if {gender[0]} participated more next semester. {name} has great stuff to say, but I never hear {gender[1]} voice or see {gender[1]} hand."
    return string

def collaboration_function(collaboration, string):
    '''adds a few sentences to the preexisting string about the student's level of collaboration skills in the class'''
    if collaboration == '5':
        string = string + f" And, {gender[1]} collaboration skills are great! When {name} is in a group, you can find {gender[2]} contributing, synthesizing, being accountable, and listening well to other group members."
    if collaboration == '4':
        string = string + f" Lucky for {gender[1]} classmates, {gender[1]} collaboration skills are quite good. In groups, {gender[0]} will be doing {gender[2]} part to make sure the project shapes up nicely."
    if collaboration == '3':
        string = string + f" Sometimes, {gender[0]} might seem apprehensive to collaborate with students {name} is less familiar with. However, once {name} gets comfortable, the work gets done."
    if collaboration == '2':
        string = string + f" Also, {gender[1]} collaborative efforts are okay, but could work on being responsible for important parts of the project. When {gender[0]} is in a group, {gender[0]} should be continue contributing, synthesizing, being accountable, and listening well to other group members."
        string = string + f" It would be great if {gender[0]} collaborated more with {gender[1]} peers next semester. There are many opportunities to improve in our upcoming group projects."
    return string

def enthusiasm_function(enthusiasm, string):
    '''adds a few sentences to the preexisting string about the student's enthusiasm level in the class'''
    if enthusiasm == '5':
        string = string + f" {name} is one of the most enthusiastic students I teach! I appreciated how {gender[0]} would always come to class ready to learn!"
    if enthusiasm == '4':
        string = string + f" {name} is very enthusiastic and {gender[0]} would usually come to class engaged and ready to learn!"
    if enthusiasm == '3':
        string = string + f" {name} is fairly enthusiastic in class and sometimes comes to class ready to learn. Next semester, I would love for {name} to meet with me if {gender[0]} might feel uninterested or unenthusiastic about what we are learning and I can take some steps to spark {gender[1]} interest once again."
    if enthusiasm == '2':
        string = string + f" One sticking point is that {name} displays a worrying lack of enthusiasm in class, often seemingly bored or uninterested in the material. I encourage {gender[2]} to meet with me in the upcoming semester to talk about ways that I can help engage {gender[2]} in class once again."
    if enthusiasm ==  '1':
        string = string + f" One sticking point is that {name} displays a worrying lack of enthusiasm in class, often seemingly bored or uninterested in the material. I encourage {gender[2]} to meet with me in the upcoming semester to talk about ways that I can help engage {gender[2]} in class once again."
    return string

def focus_function(focus, string):
    '''adds a few sentences to the preexisting string about the student's focus level in the class'''
    if focus == '5':
        string = string + f" I'm really glad {gender[0]} had such great focus this semester and {gender[0]} always got {gender[1]} work in on time and never got off task in class."
    if focus == '4': 
        string = string + f" It seemed to me that {gender[0]} had pretty good focus this semester and {gender[0]} usually got {gender[1]} work in on time and rarely got off task in class."
    if focus == '3': 
        string = string + f" From my perspective, {gender[0]} had decent focus this semester, but {gender[0]} had a few late assignments and sometimes got off task in class."
    if focus == '2': 
        string = string + f" Unfortunately, {gender[0]} rarely focused this semester. Usually off task, {gender[0]} turned in most of {gender[1]} assignments late."
    if focus == '1':
        string = string + f" Unfortunately, {gender[0]} never focused this semester. Always off task, {gender[0]} never turned in {gender[1]} assignments on time."
    return string

def loveforthecode_function(loveforthecode, string):
    '''adds a few sentences to the preexisting string about the student's love for the code in the class'''
    if loveforthecode == '5':
        string = string + f" {name} has an extremely high level of love for the code! Sometimes, I worry that it rivals my own love for the code, but then I go home and code until the sun rises and once again I am confident in my own level of love for the code. However, {gender[0]} puts up some strong competition! Every day, {gender[0]} comes bounding into class with a look of pure joy and excitement on {gender[1]} face. This type of love for the code is why I teach. I truly applaud {gender[2]}. Love. For. The. Code. This is the key to success in CS. I foresee {gender[2]} at Google or Apple in a few short years. Never stop coding, {name}. You will go far."
    if loveforthecode == '4':
        string = string + f" What a strong, loving, coder! The passion for Python is simply squeezing out of {gender[1]} every pore. {name}'s fingers fly across the keyboard like Beethoven's on the piano. Despite a few bugs here and there, {gender[0]} coding cannot be stopped due to the deep commitment underpinning it. Love. For. The. Code. This is the key to success in CS. I foresee {gender[2]} at Google or Apple, or even hacking an election in a few short years. Never stop coding, {name}. You will go far."
    if loveforthecode == '3': 
        string = string + f" But most importantly in my book, a mediocre love for the code never got anyone anywhere. Do you think that Jobs and Wozniack just created a tech conglomerate out of thin air? I certainly don't. The basis of all computer science developments EVER was love for the code. {name}, I see you trying your best. Please try and love the code just a little more. It only takes one great function or fixed bug to inspire a loving coder until the end of time."
    if loveforthecode == '2':
        string = string + f" {name}'s lack of love for the code makes me question my role as a teacher. If I cannot serve to inspire and educate, then what am I doing it all for? For my own sanity, I now ask {gender[1]} to please, please drag up the small bit of excitement that they have for the great language of Python so I can go to sleep without tears streaming down the side of my face. Even my computer keys are crusted with salt from my tears. I'm getting error warnings: your hard drive is corrupted: with sadness. I hope {name} can save my computer by simply making time to give the code the love it deserves."
    if loveforthecode == '1':
        string = string + f" I would highly suggest for {name} to drop this class as they simply don't have the amount of love for the code required to succeed in this class. Every night, I go home and wonder to myself: where have I gone wrong? I truly, truly, from the bottom of my heart, apologize if my teaching style does not inspire {gender[2]}; the only thing in life that would grant me eternal happiness is to spread my love for the code as if it is the world's most contagious disease. My partner cannot sleep at night over the sounds of my muffled, yet disappointed, cries, so, from both of us, we deplore {name} to please try and love the code a little more."
    return string
```

```python
# in this cell, we opened teachers.csv (created above) and made a file for each student named as {name}.txt
# we created a preliminary string that described the course to start out each student's file and then called the functions above to add to the string,
# eventually making a full comment for each student in their individual text file
# also, it would be easy for another teacher to use this code by changing the csv file above to reflect their students' grades and then the comment
# will print for each student 
with open('teachers.csv') as file1:
    file = csv.reader(file1, delimiter=',')
    next(file)
    for row in file:
        filename = f"{row[0]}.txt"
        studentfile = open(filename, 'w')
        name = row[0].split(' ')[0]
        final_grade = int(row[1])
        project_grade = int(row[2])
        participation = row[3]
        collaboration = row[4]
        focus = row[5]
        enthusiasm = row[6]
        loveforthecode = row[7]
        if row[8] == 'g':
            gender = ['she', 'her', 'her']
        if row[8] == 'b':
            gender = ['he', 'his', 'him']
        if row[8] == 'nb':
            gender = ['they', 'their', 'them']
        string = '    Through a project-based approach, students will work towards mastering the fundamentals of computational thinking and problem-solving using the Python programming language. A successful student will know Python syntax, data types, program organization, algorithm design, and control structures and will be able to develop efficient, readable algorithms and debug code when necessary.\n\n' 
        end_string = f' All that being said, next semester, I am excited to continue working with {name} and hope to see success and hard work during the remainder of the year. '
        string = final_grade_function(final_grade, string)
        string = project_grade_function(project_grade, string)
        string = participation_function(participation, string)
        string = collaboration_function(collaboration, string)
        string = enthusiasm_function(enthusiasm, string)
        string = focus_function(focus, string)
        string = loveforthecode_function(loveforthecode, string)
        string = string + end_string
        studentfile.write(string)
        studentfile.close()
```
